```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
from scipy import signal
from scipy.fftpack import fft, fftshift
from numpy import linalg
from scipy.stats import uniform, norm
from scipy.stats import multivariate_normal
from random import seed
from random import random 
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def RMSE(predict, true):
    return np.sqrt(np.mean(np.square(predict-true))/2)
```

```python
def plot_error_curve(train_cost_history,test_cost_history):
    plt.figure(figsize=(12, 6))
    plt.plot(train_cost_history,label='train')
    plt.plot(test_cost_history,label='test')
    plt.legend()
    plt.xlabel('EPOCH')
    plt.ylabel('RMSE')
    plt.show()
```

```python
class NeuralNetwork():
    def __init__(self, input_layer, hidden_layers1, hidden_layers2, output_layer,seed, learning_rate):
        
        self.input_layer = input_layer
        self.hidden_layers1 = hidden_layers1
        self.hidden_layers2 = hidden_layers2
        self.output_layer = output_layer
        

        np.random.seed(seed)
        
        ##### Initialize Weights 
        self.weights_input_to_hidden = np.random.normal(0.0, np.sqrt(self.input_layer), 
                                       (self.input_layer, self.hidden_layers1))
        
        self.weights_hidden_to_hidden = np.random.normal(0.0, np.sqrt(self.hidden_layers1), 
                                       (self.hidden_layers1, self.hidden_layers2))

        self.weights_hidden_to_output = np.random.normal(0.0, np.sqrt(self.hidden_layers2), 
                                       (self.hidden_layers2, self.output_layer))
        
        #####   Initialize Bias
        self.b0 = np.random.normal(0.0,np.sqrt(self.input_layer))
        self.b1 = np.random.normal(0.0,np.sqrt(self.hidden_layers1))
        self.b2 = np.random.normal(0.0,np.sqrt(self.hidden_layers2))
        
        
        self.lr = learning_rate
        
     
        self.activation_function = lambda x : 1 / (1 + np.exp(-x))
        
    
                    
    
    def train(self, features, targets):
        
        n_records = features.shape[0]
        delta_weights_i_h = np.zeros(self.weights_input_to_hidden.shape)
        delta_weights_h_h = np.zeros(self.weights_hidden_to_hidden.shape)
        delta_weights_h_o = np.zeros(self.weights_hidden_to_output.shape)
        delta_b0 = 0
        delta_b1 = 0
        delta_b2 = 0
        
        for X, y in zip(features, targets):
            
            ########Forward
            hidden_inputs = np.dot(X, self.weights_input_to_hidden)+self.b0 
            hidden_outputs = self.activation_function(hidden_inputs) 

            hidden_inputs2 = np.dot(hidden_inputs, self.weights_hidden_to_hidden)+self.b1
            hidden_outputs2 = self.activation_function(hidden_inputs2) 
          

            final_inputs = np.dot(hidden_outputs2, self.weights_hidden_to_output)+self.b2 
            final_outputs = final_inputs 
            
           
            #####Backpropogation
            error = y - final_outputs 
            output_error_term = error
            
            
            hidden_error2 = np.dot(output_error_term, self.weights_hidden_to_output.T)
            hidden_error_term2 = hidden_error2 * hidden_outputs2 * (1 - hidden_outputs2)
            
            
            hidden_error = np.dot(hidden_error_term2, self.weights_hidden_to_hidden.T)
            hidden_error_term = hidden_error * hidden_outputs * (1 - hidden_outputs)
            
            
            output_error_term = output_error_term
            hidden_error_term = hidden_error_term
            
            
            delta_weights_i_h += hidden_error_term * X[:, None]
            delta_b0 += hidden_error_term
            delta_weights_h_h += hidden_error_term2 * hidden_outputs[:, None]
            delta_b1 += hidden_error_term2
            delta_weights_h_o += output_error_term * hidden_outputs2[:, None]
            delta_b2 += output_error_term

        
        self.weights_hidden_to_output += self.lr * delta_weights_h_o / n_records 
        self.b2 += self.lr*delta_b2/n_records
        
        self.weights_hidden_to_hidden += self.lr * delta_weights_h_h / n_records 
        self.b1 += self.lr*delta_b1/n_records
        
        self.weights_input_to_hidden += self.lr * delta_weights_i_h / n_records 
        self.b0 += self.lr*delta_b0/n_records
        
        return self.weights_hidden_to_output,self.weights_hidden_to_hidden,self.weights_input_to_hidden,self.b2,self.b1,self.b0
```

```python
def test(data,weights_h2o,weights_h2h,weights_i2h,b2,b1,b0):
    predicts = np.zeros(data.shape[0])
    for i in range(data.shape[0]):
        X = data[i]
        hidden_inputs = np.dot(X, weights_i2h)+b0 
        hidden_outputs = sigmoid(hidden_inputs) 

        hidden_inputs2 = np.dot(hidden_inputs, weights_h2h)+b1
        hidden_outputs2 = sigmoid(hidden_inputs2) 
          

        final_inputs = np.dot(hidden_outputs2, weights_h2o)+b2 
        final_outputs = final_inputs 
        #print(X)
        predicts[i] = final_outputs
        
    return predicts
```

```python
def batch_gradient_descent(train_x,train_y,test_x,test_y,iterations=100):
    m = len(train_y)
    train_cost_history = np.zeros(iterations)
    test_cost_history = np.zeros(iterations)
    for i in range(iterations):
        (w_h2o,w_h2h,w_i2h,b2,b1,b0) = network.train(train_x,train_y)
        predict_train = test(train_x,w_h2o,w_h2h,w_i2h,b2,b1,b0)
        #predict_train = predict_train.reshape([-1])
        train_cost = RMSE(predict_train,train_y)
        train_cost_history[i]  = train_cost
        
        
        predict_test = test(test_x,w_h2o,w_h2h,w_i2h,b2,b1,b0)
        #predict_test = predict_test.reshape([-1])
        test_cost = RMSE(predict_test,test_y)
        test_cost_history[i]  = test_cost
    return train_cost_history,test_cost_history
```

```python
axis_num = 50 ####number of datapoint in each dimension
data_num = axis_num**2  ####Total number data point 
train_num = int(data_num*0.8) ######Number of training set
```

```python
range1 = np.linspace(-1,1,axis_num)   #####Variable range
range2 = np.linspace(-1,1,axis_num)   ####Variable Range
```

```python
x1 = np.zeros(data_num)      #####Varaible1 
x2 = np.zeros(data_num)      #####Variable2
```

```python
for i in range(data_num):
    x1[i] = range1[i%axis_num]
    x2[i] = range2[i//axis_num]
```

```python
noise = norm.rvs(loc=0, scale=0.1, size=(data_num,))    ####Gausian Noise
```

```python
y = x1+x2**2+(2*x1*np.cos(x2))**2+noise   #f(x)
```

```python
inputs = np.array([x1,x2]).T
```

```python
permutation = np.random.permutation(y.shape[0])
inputs = inputs[permutation]
y = y[permutation]
train_x = inputs[0:train_num,]
train_y = y[0:train_num]
test_x = inputs[train_num:]
test_y = y[train_num:]
```

```python
input_layer=2  ####Node number of input layer 
hidden_layer1 = 10   ####Node number of hidden layer 1
hidden_layer2 = 20    ####Node number of hidden layer 2
output_layer = 1     ####Node number of output layer
seed = 0            ####seed of random for weights
learning_rate = 0.01  ####Leraning rate
```

```python
network = NeuralNetwork(input_layer,hidden_layer1,hidden_layer2,output_layer,seed, learning_rate)
```

```python
train_cost_history,test_cost_history = batch_gradient_descent(train_x,train_y,test_x,test_y)
```
