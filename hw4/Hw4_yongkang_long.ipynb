```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import statistics
```

```python
# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
df = pd.read_csv('clu_data.txt', delimiter=' ', header=None)
```

```python
df.columns = ['x', 'y']
```

```python
plt.scatter(df['x'], df['y'])
```

```python
## Assignment Stage
def assignment(df, centroids):
    for i in centroids.keys():
        
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x'] - centroids[i][0]) ** 2
                + (df['y'] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    #df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

## Update Stage
def update(centroids):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['x'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['y'])
    return centroids

def calculate_sse(df):
    sse = 0.0
    for i in range(1000):
        sse += df['distance_from_'+str(df['closest'][i])][i]
    return sse
        

def plot(data):
    fig = plt.figure(figsize=(5, 5))
    plt.scatter(data['x'], data['y'], color='k')
    colmap = {1: 'r', 2: 'g', 3: 'b',4:'y', 5: 'c', 6: 'm', 7:'k'}
    #for i in centroids.keys():
     #   plt.scatter(*centroids[i], color=colmap[i])
    plt.show()
```

```python

#####Random Centroids
def kmeans(k,df):
    centroids = {
        i+1: [np.random.randint(-25, 15), np.random.randint(-15, 15)]
        for i in range(k)
    }
    df = assignment(df, centroids)
# Continue until all assigned categories don't change any more
    while True:
        closest_centroids = df['closest'].copy(deep=True)
        centroids = update(centroids)
        df = assignment(df, centroids)
        if closest_centroids.equals(df['closest']):
            break
    return df
    
```

```python
np.random.seed(200)
k_list = []
sse_list = []
for k in range(1,30):
    kmeans(k,df)
    sse = calculate_sse(df)
    k_list.append(k)
    sse_list.append(sse)
```

# 1) Try different settings of parameter k. For each value of k, compute the SSE (sum of squared error) of clustering result. Plot the SSE curve w.r.t. the various k values. 

```python
plt.title('SSE curve')
plt.xlabel('Number of Cluster')
plt.ylabel('With Groups SSE')
plt.plot(k_list,sse_list)
```
