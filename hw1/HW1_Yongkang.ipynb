## Randomly choose 80% of the data for learning the regression function. Use the remaining 20% for evaluating the learned function.

```python
df = shuffle(df)
train_num = int(df.shape[0]*ratio)
train=df.iloc[:train_num]
test = df.iloc[train_num:]
train_x,train_y = np.array(train['X'],dtype='float64'),np.array(train['Y'],dtype='float64')
test_x,test_y = np.array(test['X'],dtype='float64'),np.array(test['Y'],dtype='float64')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train_x,train_y,'o')
plt.title("Train Actual Point Distribution")
```

```python
def polynomial_basic_function(x,degree=2):
    power = np.ones((x.shape[0],degree))*np.arange(0,degree,1)
    new_x = np.power(x,power.T,dtype='float64')
    return new_x.T
```

```python
def plot_error_curve(train_cost_history,test_cost_history,xlabel):
    plt.figure(figsize=(12, 6))
    plt.plot(train_cost_history,label='train')
    plt.plot(test_cost_history,label='test')
    plt.legend()
    plt.xlabel(xlabel)
    plt.ylabel('MSE')
    plt.show()
    
```

# (1)	(3 points) implement the batch gradient descent algorithm. 

```python
def batch_gradient_descent(train_x,train_y,test_x,test_y,w,learning_rate=0.01,iterations=20):
    m = len(train_y)
    w_history = np.zeros((iterations,degree))
    train_cost_history = np.zeros(iterations)
    test_cost_history = np.zeros(iterations)
    for i in range(iterations):
        predict_y = np.matmul(train_x, w)
        train_cost = 1/(2*m) * np.sum(np.square(predict_y-train_y))
        train_cost_history[i]  = train_cost
        
        w = w -(1/m)*learning_rate*np.matmul(train_x.T,predict_y - train_y)
        w_history[i,:] =w.T
        
        predict_test = np.matmul(test_x, w)
        test_cost = 1/(2*len(test_y)) * np.sum(np.square(predict_test-test_y))
        test_cost_history[i]  = test_cost
    return w, train_cost_history, test_cost_history,w_history
```

```python
degree = 2
new_train_x = polynomial_basic_function(train_x,degree)
new_test_x = polynomial_basic_function(test_x,degree)
w = np.ones(degree)
```

```python
w,train_cost_history,test_cost_history,w_history = batch_gradient_descent(new_train_x,train_y,new_test_x,test_y,w,0.01,100)
```

# (2)	  (3 points) implement the stochastic gradient descent algorithm. 

```python
def stochastic_gradient_descent(train_x,train_y,test_x,test_y,w,learning_rate=0.01,iterations=10):
    m = len(train_y)
    w_history = np.zeros((iterations,degree))
    train_cost_history = np.zeros(iterations)
    test_cost_history = np.zeros(iterations)
    train_cost_datapoint = np.zeros(iterations*m)
    k = 0
    for i in range(iterations):
        train_cost = 0.0
        for j in range(m):
            rand_ind = np.random.randint(0,m)
            #rand_ind = j
            X_i = train_x[rand_ind,:].reshape(1,train_x.shape[1])
            y_i = train_y[rand_ind].reshape(1)
            predict_yi = np.matmul(X_i,w)
            #w = w -(1/m)*learning_rate*np.matmul(train_x.T,predict_y - train_y)
            w = w -(1/len(y_i))*learning_rate*(np.matmul(X_i.T,(predict_yi - y_i)))
            cost = 1/(2*len(y_i)) * np.sum(np.square(predict_yi-y_i))
            train_cost_datapoint[k] = cost
            train_cost += cost
            k = k+1
            
        train_cost_history[i]  = train_cost/m
        w_history[i,:] =w.T
        predict_test = np.matmul(test_x, w)
        test_cost = 1/(2*len(test_y)) * np.sum(np.square(predict_test-test_y))
        test_cost_history[i]  = test_cost
    return w,train_cost_history, test_cost_history,w_history,train_cost_datapoint
```
